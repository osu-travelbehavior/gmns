```python
net = og.getNetFromFile('SourceData/ohio_columbus_15.pbf', network_types=('auto'), connector_link_types=['unclassified'], POI=True, POI_sampling_ratio=1.0)
og.consolidateComplexIntersections(net, auto_identify=True)
og.generateNodeActivityInfo(net)

og.outputNetToCSV(net, output_folder = 'output_col/auto')
```

```python
import path4gmns as pg
import os
import pandas as pd

path = '/Users/kong/Library/CloudStorage/OneDrive-TheOhioStateUniversity/data/'
file_name = 'od_gmns_node_join'
od_list = pd.read_csv(path+file_name+'.csv')
od_list.loc[:,['route']] = ''

os.chdir('/Users/kong/Library/CloudStorage/OneDrive-TheOhioStateUniversity/code/output_col/auto')
network = pg.read_network()

for i in range(len(od_list)) :
    ori = int(od_list.loc[i, ['node_id']][0])
    des = int(od_list.loc[i, ['work_node_id']][0])

    output = network.find_shortest_path(str(ori), str(des), seq_type='link')
    od_list.loc[i, ['route']] = output
    print(str(i+1)+'/'+str(len(od_list)))

od_list.to_csv(path+file_name+'_output'+'.csv', header=True, index=False)
```
